# **Outlier Treatment**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1_columns=['LB','AC','FM','UC','DL','DS','DP','ASTV','MLTV','Width','Tendency','NSP']
for i in df1_columns:
  df1[i].plot(kind='box')
  plt.title(f'Box plot of {i}') # Add a title for clarity
  plt.show() # Display each plot
```

```python
df1_columns=['LB','AC','UC','DL','ASTV','MLTV','Width','Tendency']

for col in df1_columns:
  Q1 = df1[col].quantile(0.25)
  Q3 = df1[col].quantile(0.75)
  IQR = Q3 - Q1
  UL = Q3 + 1.5 * IQR
  LL = Q1 - 1.5 * IQR

  # Cap the outliers
  df1.loc[df1[col] > UL, col] = UL
  df1.loc[df1[col] < LL, col] = LL

  print(f"Outliers treated for column: {col}")
```

```python
df1_columns=['LB','AC','UC','DL','ASTV','MLTV','Width','Tendency']
for i in df1_columns:
  df1[i].plot(kind='box')
  plt.title(f'Box plot of {i}') # Add a title for clarity
  plt.show() # Display each plot
```
